```python
# data ingestion in batch due to large number of data
df = pd.read_csv('green_tripdata_2019-10.csv', nrows=100)
```

```python
df
```

```python
# parse tpep* converted from text to timestamp
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
# python library to connect postgres to python
from sqlalchemy import create_engine
```

```python
#'postgresql://dbusername:dbpassword@localhost:host_portnumber/db_name'
engine = create_engine('postgresql://root:root@localhost:5433/ny_taxi')
```

```python
# print schema for DDL (Data Definition Language) for postgres
print(pd.io.sql.get_schema(df, name='green_taxi_data', con=engine))
```

```python
# iteration of data ingestion in batch due to large number of data
df_iter = pd.read_csv('green_tripdata_2019-10.csv', iterator=True, chunksize=100000)
```

```python
# parse tpep* converted from text to timestamp
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
df
```

```python
# create header in sql
df.head(n=0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
```

```python
# insert first batch of data into postgres and record time taken to finish
%time df.to_sql(name='green_taxi_data', con=engine, if_exists='append')
```

```python
from time import time
```

```python
while True:
    t_start = time()

    df = next(df_iter)
    
    df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)

    df.to_sql(name='green_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print('insert another chunk..., time taken %.3f seconds' % (t_end - t_start))
```
